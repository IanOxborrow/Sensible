```python
# import libraries
import pandas as pd
from sklearn import preprocessing
import sqlite3
import pandas as pd
import math
import numpy as np



da = pd.read_csv('accelerometer/pendulum/slide/phyphox.csv')  
db = pd.read_csv('accelerometer/pendulum/slide/testproject.csv')
# dc = table for accelerometer
dd = pd.read_csv('gyruscope/phyphox.csv')  
de = pd.read_csv('gyruscope/testproject.csv')  
# dg = table for the gyruscope
dh = pd.read_csv('microphone/phyphox.csv')  
di = pd.read_csv('microphone/testproject.csv') 
# dj = table for the microphone
dj = pd.read_csv('gravity test/phyphox.csv')  
dk = pd.read_csv('gravity test/testproject.csv') 
```

```python
# accelerometer Accuracy testing with sample rate by default
# 5 seconds of moving the phone in the same direction up,down,left,right



# Combine two tables
dc = pd.concat([da,db], axis = 1)
dc = dc.drop(columns = ['d']) #droping nan column
dc = dc.dropna() # droping nan rows
# dy = dc.values
# min_max_scaler = preprocessing.MinMaxScaler()
# x_scaled = min_max_scaler.fit_transform(dy)
# dc = pd.DataFrame(x_scaled)


# accuracy = (Value from phyphox - Value from Testproject) / Value from phyphox * 100%
accuracy_x = abs((da['Acceleration x (m/s^2)'] - (db['Acceleration x (m/s^2)']))/da['Acceleration x (m/s^2)'])*100
accuracy_y = abs((da['Acceleration y (m/s^2)'] - (db['Acceleration y (m/s^2)']))/da['Acceleration y (m/s^2)'])*100
accuracy_z = abs((da['Acceleration z (m/s^2)'] - (db['Acceleration z (m/s^2)']))/da['Acceleration z (m/s^2)'])*100

# difference = expected value - actual value
difference_x = da['Acceleration x (m/s^2)'] - (db['Acceleration x (m/s^2)'])
difference_y = da['Acceleration y (m/s^2)'] - (db['Acceleration y (m/s^2)'])
difference_z = da['Acceleration z (m/s^2)'] - (db['Acceleration z (m/s^2)'])

dc['Accuracy of acceleration x (%)'] = accuracy_x
dc['Accuracy of acceleration y (%)'] = accuracy_y
dc['Accuracy of acceleration z (%)'] = accuracy_z

dc['Difference between expected value and actual value on x'] = difference_x
dc['Difference between expected value and actual value on y'] = difference_y
dc['Difference between expected value and actual value on z'] = difference_z

dc.replace([np.inf, -np.inf], np.nan, inplace=True)
dc = dc.dropna() # droping nan rows

# average percentage change and #average difference between expected value and actual value
dc = dc.append({'Accuracy of acceleration x (%)' : (dc['Accuracy of acceleration x (%)'].sum()/len(dc['Accuracy of acceleration x (%)']))
                , 'Accuracy of acceleration y (%)' : (dc['Accuracy of acceleration y (%)'].sum()/len(dc['Accuracy of acceleration y (%)'])), 
                'Accuracy of acceleration z (%)' : (dc['Accuracy of acceleration z (%)'].sum()/len(dc['Accuracy of acceleration z (%)'])), 
               'Difference between expected value and actual value on x' : (dc['Difference between expected value and actual value on x'].sum()/len(dc['Difference between expected value and actual value on x']))
                , 'Difference between expected value and actual value on y' : (dc['Difference between expected value and actual value on y'].sum()/len(dc['Difference between expected value and actual value on y'])), 
                'Difference between expected value and actual value on z' : (dc['Difference between expected value and actual value on z'].sum()/len(dc['Difference between expected value and actual value on z']))},ignore_index = True)

dc

# closer to 0 %, more accurate
```

```python



# accelerometer Accuracy testing with sample rate by default
# 5 seconds of moving the phone in the same direction up,down,left,right



# Combine two tables
dc = pd.concat([da,db], axis = 1)
dc = dc.drop(columns = ['d']) #droping nan column
dc = dc.dropna() # droping nan rows
dy = dc.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(dy)
dc = pd.DataFrame(x_scaled)

accuracy_x = abs(abs(dc[1]-dc[5])/dc[1])*100
accuracy_y = abs(abs(dc[2]-dc[6])/dc[2])*100
accuracy_z = abs(abs(dc[3]-dc[7])/dc[3])*100
dc['Accuracy of acceleration x (%)'] = accuracy_x
dc['Accuracy of acceleration y (%)'] = accuracy_y
dc['Accuracy of acceleration z (%)'] = accuracy_z
dc.replace([np.inf, -np.inf], np.nan, inplace=True)
dc = dc.dropna() # droping nan rows
dc = dc.append({'Accuracy of acceleration x (%)' : (dc['Accuracy of acceleration x (%)'].sum()/len(dc['Accuracy of acceleration x (%)']))
                , 'Accuracy of acceleration y (%)' : (dc['Accuracy of acceleration y (%)'].sum()/len(dc['Accuracy of acceleration y (%)'])), 
                'Accuracy of acceleration z (%)' : (dc['Accuracy of acceleration z (%)'].sum()/len(dc['Accuracy of acceleration z (%)']))},ignore_index=True)

dc
# closer to 0 %, more accurate
```
